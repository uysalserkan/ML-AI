```python
import tensorflow as tf
import tensorflow_hub as hub
import matplotlib.pyplot as plt
import numpy as np
```

```python
content = plt.imread('ibo.jpeg')
```

```python
texture = plt.imread('pika2.jpeg')
```

```python
content = content.astype(np.float32)[np.newaxis, ...] / 255.
texture = texture.astype(np.float32)[np.newaxis, ...] / 255.
```

```python
texture = tf.image.resize(texture, (256, 256))
```

```python
module = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
outputs = module(tf.constant(content), tf.constant(texture))
stylized_image = outputs[0]
```

```python
plt.imshow(stylized_image[0])
```

```python
plt.imsave("output16.jpg", np.asarray(stylized_image[0]))
```
